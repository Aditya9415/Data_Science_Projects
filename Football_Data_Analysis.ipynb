```python
player_salary=fifa[['short_name','wage_eur']]
```

```python
player_salary.head()
```

```python
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
player_salary.head()
```

## Plotting player earning

```python
plt.figure(figsize=(8,5))
plt.bar(player_salary['short_name'][0:5],player_salary['wage_eur'][0:5],color=['red','blue','green','pink','orange'])
plt.show()
```
